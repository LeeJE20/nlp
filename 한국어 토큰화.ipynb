```python
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Komoran

okt = Okt()
kkma = Kkma()
komoran = Komoran()
```

```python
sentence = ["나는 하늘을 나는 비행기를 좋아해서 열심히 일하고 난 뒤 주ㅜ말에 여행을 떠날거야ㅎ ㅎ", "흐ㅎㅎㅎㅎㅎㅎㅎㅎㅎ난느ㅡㄴ 오타어ㅔ 강ㅇ한 문자ㅏㅇ이ㅑㅋㅋㅋ쿠ㅜㅜㅜ", "외래어 문장: 이 옷은 소재가 린넨이라 여름에 시원하고 컬러로는 카멜, 아이보리, 블랙, 화이트, 스카이 블루가 있어요. 퍼스널 컬러가 여름 쿨 브라이트면 잘 어울려요.", "노을이 진 하늘 끝에 걸려있는 구름 한 조각을 퍼담아 왔어"]
```

```python
for s in sentence:
    print(f"\nsentence: {s}\n\n")
    
    print("[okt]")
    print('형태소 분석 :',okt.morphs(s))
    print('품사 태깅 :',okt.pos(s))
    print('명사 추출 :',okt.nouns(s)) 
    print()
    
    print("[kkma]")
    print('형태소 분석 :',kkma.morphs(s))
    print('품사 태깅 :',kkma.pos(s))
    print('명사 추출 :',kkma.nouns(s)) 
    print()
    
    print("[komoran]")
    print('형태소 분석 :',komoran.morphs(s))
    print('품사 태깅 :',komoran.pos(s))
    print('명사 추출 :',komoran.nouns(s)) 
    print()
```
